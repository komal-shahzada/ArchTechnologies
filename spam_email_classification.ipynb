```python
# Importing necessary libraries for data processing and machine learning
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Libraries imported successfully!")
```

```python
# Load the dataset (Make sure you have a csv file or use this sample data)
# Example: df = pd.read_csv('spam.csv', encoding='latin-1')

data = {
    'text': [
        'Free entry in 2 a weekly comp to win FA Cup final tkts 21st May 2005.',
        'Hi, how are you doing today?',
        'WINNER!! As a valued network customer you have been selected to receivea £900 prize reward!',
        'Can we meet for coffee tomorrow at 10?',
        'URGENT! You have won a 1 week FREE membership in our £100,000 Prize Jackpot!',
        'I will call you later, I am in a meeting right now.'
    ],
    'label': ['spam', 'ham', 'spam', 'ham', 'spam', 'ham']
}

df = pd.DataFrame(data)

# Mapping labels to numbers: ham = 0, spam = 1
df['label_num'] = df['label'].map({'ham': 0, 'spam': 1})

print(df.head())
```

```python
# Splitting the data into training and testing sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label_num'], test_size=0.2, random_state=42)

print(f"Training samples: {len(X_train)}")
print(f"Testing samples: {len(X_test)}")
```

```python
# Initializing CountVectorizer to convert text data into numerical vectors
vectorizer = CountVectorizer()

# Fitting and transforming the training data
X_train_count = vectorizer.fit_transform(X_train.values)

print("Text data converted to numerical format.")
```

```python
# Initializing the Naive Bayes model
model = MultinomialNB()

# Training the model with our training data
model.fit(X_train_count, y_train)

print("Model training completed!")
```

```python
# Transforming the test data and making predictions
X_test_count = vectorizer.transform(X_test)
predictions = model.predict(X_test_count)

# Printing the accuracy and performance report
print("Accuracy Score:", accuracy_score(y_test, predictions))
print("\nClassification Report:\n", classification_report(y_test, predictions))
```

```python
# Install gradio if you haven't already
# !pip install gradio

import gradio as gr

print("Gradio library imported successfully!")
```

```python
# Function to predict if the input email is Spam or Ham
def predict_email(email_text):
    # Transforming input text using the previously trained vectorizer
    email_count = vectorizer.transform([email_text])

    # Making prediction
    prediction = model.predict(email_count)

    # Returning the result as a string
    if prediction[0] == 1:
        return "⚠️ SPAM ALERT!"
    else:
        return "✅ This is a HAM (Safe) Email."

print("Prediction function is ready.")
```

```python
# Creating the Gradio Interface
interface = gr.Interface(
    fn=predict_email,
    inputs=gr.Textbox(lines=5, placeholder="Enter the email content here...", label="Email Text"),
    outputs=gr.Label(label="Result"),
    title="Email Spam Classifier - Task 1",
    description="Internship Project for Arch Technologies. Enter any email text below to check if it's Spam or Ham.",
    theme="soft"
)

# Launching the interface
interface.launch(share=True)
```
